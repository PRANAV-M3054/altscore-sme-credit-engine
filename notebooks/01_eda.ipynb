```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 100)
pd.set_option('display.float_format','{:.2f}'.format)
```

```python
app = pd.read_csv('../data/raw/application_data.csv')
print("Shape:",app.shape)
print("\nFirst 5 rows:")
app.head()
```

```python
print("Columns:", app.columns.tolist())
```

```python
missing = app.isnull().sum()
missing_pct =(missing / len(app) * 100).round(2)
missing_df =pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
}).query('missing_count > 0').sort_values('missing_pct', ascending=False)

print(f"Columns with missing values: {len(missing_df)}")
print(missing_df.head(20))
```

```python
print("Target  distribution (TARGET column):")
print(app['TARGET'].value_counts())
print("\nDefault rate:", round(app["TARGET"].mean() *100,2),"%")

```

```python
key_cols = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY',
            'AMT_GOODS_PRICE', 'DAYS_BIRTH', 'DAYS_EMPLOYED',
            'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']

print(app[key_cols].describe().round(2))
```

```python
print("Income types:")
print(app["NAME_INCOME_TYPE"].value_counts())

print("\nOrganisation types (top 15):")
print(app["ORGANIZATION_TYPE"].value_counts().head(15))
```
