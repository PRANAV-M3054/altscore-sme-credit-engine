```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 100)
pd.set_option('display.float_format','{:.2f}'.format)
```

```python
app = pd.read_csv('../data/raw/application_data.csv')
print("Shape:",app.shape)
print("\nFirst 5 rows:")
app.head()
```

```python
print("Columns:", app.columns.tolist())
```

```python
missing = app.isnull().sum()
missing_pct =(missing / len(app) * 100).round(2)
missing_df =pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
}).query('missing_count > 0').sort_values('missing_pct', ascending=False)

print(f"Columns with missing values: {len(missing_df)}")
print(missing_df.head(20))
```

```python
print("Target  distribution (TARGET column):")
print(app['TARGET'].value_counts())
print("\nDefault rate:", round(app["TARGET"].mean() *100,2),"%")

```

```python
key_cols = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY',
            'AMT_GOODS_PRICE', 'DAYS_BIRTH', 'DAYS_EMPLOYED',
            'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']

print(app[key_cols].describe().round(2))
```

```python
print("Income types:")
print(app["NAME_INCOME_TYPE"].value_counts())

print("\nOrganisation types (top 15):")
print(app["ORGANIZATION_TYPE"].value_counts().head(15))
```

```python
df_clean = pd.read_csv('../data/processed/application_cleaned.csv')
print("Cleaned data shape:", df_clean.shape)
print("Missing values:", df_clean.isnull().sum().sum())
```

```python
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
correlation = df_clean[numeric_cols].corr()['TARGET'].sort_values()

print("Top 10 columns Most correlated with default (positive): ")
print(correlation.tail(10))

print("\nTop 10 columns LEAST correlated with default (negative): ")
print(correlation.head(10))
```

```python
import matplotlib.pyplot as plt
 
corr_values = correlation.drop('TARGET')

top_bottom = pd.concat([corr_values.head(10), corr_values.tail(10)])

plt.figure(figsize=(10,8))
colors = ['red' if x > 0 else 'green' for x in top_bottom.values]
plt.barh(top_bottom.index, top_bottom.values, color=colors)
plt.axvline(x =0, color ='black', linewidth = 0.8)
plt.title('Top 20 Features Correlated with Default')
plt.xlabel('Corelation with TARGET')
plt.tight_layout()
plt.savefig('../reports/figures/correlation_plot.png', dpi = 150)
plt.show()
print("Plot saved to reports/figures/")
```
